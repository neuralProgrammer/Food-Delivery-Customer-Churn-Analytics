```python
import os
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", None)
pd.set_option("display.width", None)
```

```python
DATA_PATH = "../data/raw"
```

```python
datasets = {}

for file in sorted(os.listdir(DATA_PATH)):
    if file.endswith(".csv"):
        path = os.path.join(DATA_PATH, file)
        datasets[file] = pd.read_csv(path)

print(f"Loaded {len(datasets)} datasets.\n")

for name in datasets:
    print(name)
```

```python
def summarize_dataset(name, df):
    print("=" * 100)
    print(f"Dataset : {name}")
    print("=" * 100)

    print(f"\nShape              : {df.shape}")
    print(f"Rows               : {df.shape[0]:,}")
    print(f"Columns            : {df.shape[1]}")

    print("\nColumns")
    print("-" * 100)
    print(df.columns.tolist())

    print("\nData Types")
    print("-" * 100)
    print(df.dtypes)

    print("\nMissing Values")
    print("-" * 100)
    missing = df.isnull().sum()
    missing = missing[missing > 0]

    if len(missing) == 0:
        print("No Missing Values")
    else:
        print(missing.sort_values(ascending=False))

    print("\nDuplicate Rows")
    print("-" * 100)
    print(df.duplicated().sum())

    print("\nMemory Usage (MB)")
    print("-" * 100)
    print(round(df.memory_usage(deep=True).sum() / 1024**2, 2))

    print("\nFirst Five Rows")
    print("-" * 100)
    display(df.head())

    print("\n")
```

```python
for name, df in datasets.items():
    summarize_dataset(name, df)
```

```python
summary = []

for name, df in datasets.items():
    summary.append({
        "Table": name,
        "Rows": df.shape[0],
        "Columns": df.shape[1],
        "Missing Values": int(df.isnull().sum().sum()),
        "Duplicate Rows": int(df.duplicated().sum())
    })

summary_df = pd.DataFrame(summary)

summary_df.sort_values("Rows", ascending=False)
```

```python
column_summary = []

for table_name, df in datasets.items():

    for column in df.columns:

        column_summary.append({

            "Table": table_name,

            "Column": column,

            "Data Type": str(df[column].dtype),

            "Missing": int(df[column].isnull().sum()),

            "Unique Values": int(df[column].nunique())

        })

column_summary = pd.DataFrame(column_summary)

column_summary.head(20)
```

```python
summary_df.to_csv("../docs/table_summary.csv", index=False)

column_summary.to_csv("../docs/column_summary.csv", index=False)

print("Documentation exported successfully.")
```
